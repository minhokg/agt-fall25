## B) Empirical illustration

### B-1) Applicant-Optimality
It means each applicant is matched to their best possible option in any stable matching.

```python
# call check_applicant_optimality.py
from ps3.functions.check_applicant_optimality import check_applicant_optimality
```

```python
# check whether the result of previous section is optimal or not
check_applicant_optimality(applicants_result=applicants_result, hospitals_result=hospitals_result, applicants_preferences=applicants_preferences)
```

```python
# Now generate random preference dataset and check whether every result are optimal or not
from ps3.functions.generate_random_preferences import generate_random_preferences
import random

# Set number of datasets for checking
n_datasets = 1000

# Set maximum number of applicants and hospitals
max_n_applicants = 10
max_n_hospitals = 10

# Set a default value for optimality. It can be True/False
optimality = False

# Run the loop for checking each case's applicant optimality
for i in range(n_datasets):
    # generate random
    num_applicants = random.randint(1, max_n_applicants)
    num_hospitals = random.randint(1, max_n_hospitals)
    applicants_preferences, hospitals_preferences = generate_random_preferences(num_applicants=num_applicants, num_hospitals=num_hospitals)
    applicants_result, hospitals_result = deferred_acceptance(applicants_preferences=applicants_preferences, hospitals_preferences=hospitals_preferences)
    optimality = check_applicant_optimality(applicants_result=applicants_result, hospitals_result=hospitals_result, applicants_preferences=applicants_preferences)
    if not optimality:
        print(f'Applicant_result: {applicants_result} \n Hospital_result: {hospitals_result} \n are not applicant-optimal')
        break

# If every case is optimal, then it prints the below
if optimality:
    print('All datasets are applicant-optimal')
```

### B-2) Is the deferred acceptance algorithm DSIC for the applicants, not for the hospitals?

```python
# call manipulate_hospitals_preferences.py
from ps3.functions.manipulate_hospitals_preferences import manipulate_hospitals_preferences
```

```python
# come back to the original example.
applicants_preferences = {
    'A': ['H2', 'H1', 'H3'],
    'B': ['H1', 'H3', 'H2'],
    'C': ['H3', 'H2', 'H1'],
}

hospitals_preferences = {
    'H1': ['A', 'B', 'C'],
    'H2': ['B', 'A', 'C'],
    'H3': ['C', 'A', 'B'],
}
```

```python
# reverse hospitals
manipulated_hospitals_preferences = manipulate_hospitals_preferences(hospitals_preferences=hospitals_preferences)
manipulated_hospitals_preferences
```

```python
# run the algorithm with truthful preferences
applicants_result, hospitals_result = deferred_acceptance(applicants_preferences=applicants_preferences, hospitals_preferences=hospitals_preferences)

# run the algorithm with manipulated hospital preferences
applicants_result_manipulation, hospitals_result_manipulation = deferred_acceptance(applicants_preferences=applicants_preferences,hospitals_preferences=manipulated_hospitals_preferences)

# check if any applicant is worse off in the manipulated scenario
applicants_worse_off = False
for applicant in applicants_preferences:
    if applicants_preferences[applicant].index(applicants_result[applicant]) > applicants_preferences[applicant].index(applicants_result_manipulation[applicant]):
        applicants_worse_off = True
        print(f"Truthful reporting doesn't work for {applicant}")
        break

if not applicants_worse_off:
    print("Telling truth is better even though hospitals manipulate their preferences")
```

## C) Used Deferred Acceptance algorithm in Job Market

Matching employers with unemployed people can be use case for deferred acceptance algorithm. There are two reasons.
1. It is an applicant-optimality. That is, each unemployed person can be matched with the best possible option. So, this is benefical for unemployed people, who are minor in the society.
2. It makes applicant to take a truth telling strategy. That is, there is no motivation for applicant to be better off by lying. So that, employers can recognize applicant's truthful preference
